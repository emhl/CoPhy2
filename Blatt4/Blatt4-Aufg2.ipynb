### a) Plotte die ersten sechs Legendre-Polynome (scipy.special.legendre(n)).

```python
# plot the first 5 legendre polynomials
x = np.linspace(-1, 1, 100)
for i in range(6):
    plt.plot(x, special.eval_legendre(i, x), label="P_{}".format(i))
plt.legend()
plt.show()
```
